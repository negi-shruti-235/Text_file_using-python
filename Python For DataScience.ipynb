```python
def ret(x):
    return contxt.count(x)
file=open("C:\conv.txt", 'r')
contxt=file.read()
names=["WILL", "ROYCE", "GARED", "JON", "SEPTA MORDANE", "SANSA", "NED", "ROBB", "CASSEL", "CATELYN", "BRAN", "THEON", "CERSEI", "JAIME", "ROBERT", "ARYA"]
td=0
print("Total Dialouge spoken by the Character:")
for i in range(16):
    name=names[i]
    print(name, ":", ret(name))
    td=td+ret(name)
print("Total dialouges spoken in the conversation are:", td)
file.close()
```

```python
file=open("C:\conv.txt","r")
file1=open("Will.txt","w")
for i in range(223):
    line=file.readline()
    if "WILL" in line:
        line=line.lstrip("WILL:")
        new=line.split()
        for i in new:
            file1.write(i+"\n")          
file.close()
file1.close()
file1=open("Will.txt", "r")
print(file1.read())
file1.close()
```
